Url = https://archive.ics.uci.edu/ You have to find the following details: A) Dataset name B) Data type C) Task D) Attribute type E) No of instances F) No of attribute G) Year

```python
driver=webdriver.Chrome('chromedriver.exe')
url='https://archive.ics.uci.edu/'
driver.get(url)
```

```python
all_dataset=driver.find_elements_by_xpath("//span[@class='normal']//a")
for i in all_dataset:
    if i.text=='view all data sets':
        i.click()
        break
```

```python

Dataset_Name=[]

name=driver.find_elements_by_xpath("//table[@cellpadding=5]/tbody/tr/td[1]")

for i in name:
    if i.text is None:
        Dataset_Name.append("-")
    else:
        Dataset_Name.append(i.text)
    
len(Dataset_Name)
```

```python
Data_Type=[]

dt=driver.find_elements_by_xpath("//table[@cellpadding=5]/tbody/tr/td[2]")

for i in dt:
    if i.text is None:
        Data_Type.append("-")
    else:
        Data_Type.append(i.text)
    
len(Data_Type)
```

```python
Default_Task=[]

det=driver.find_elements_by_xpath("//table[@cellpadding=5]/tbody/tr/td[3]")

for i in det:
    if i.text is None:
        Default_Task.append("-")
    else:
        Default_Task.append(i.text)
    
len(Default_Task)
```

```python
Attributes_Types=[]  #list

at=driver.find_elements_by_xpath("//table[@cellpadding=5]/tbody/tr/td[4]") #xapth for attributes type

for i in at:
    if i.text is None: #loop
        Attributes_Types.append("-")
    else:
        Attributes_Types.append(i.text)
    
len(Attributes_Types)  #Checking the length of attributes type
```

```python
Instances=[] #List

ins=driver.find_elements_by_xpath("//table[@cellpadding=5]/tbody/tr/td[5]") #xpath for instances

for i in ins:  #loop
    if i.text is None:
        Instances.append("-")
    else:
        Instances.append(i.text)
    
len(Instances)  # Checking the length of instances
```

```python
Attributes=[]   #list

atr=driver.find_elements_by_xpath("//table[@cellpadding=5]/tbody/tr/td[6]")  #find xpath of the given attributes

for i in atr:   # loop
    if i.text is None:
        Attributes.append("-")
    else:
        Attributes.append(i.text)
        
len(Attributes)  #Checking the length of the attributes
```

```python
Attributes
```

```python
Year=[]  #list 

yr=driver.find_elements_by_xpath("//table[@cellpadding=5]/tbody/tr/td[7]")  #xpath

for i in yr:  #loop
    try:
        Year.append(i.text)
    except NoSuchElementException:
        Year.append("--")
        
len(Year)  #checking length of the year
```

```python

Year
```

```python
#Creating data frame
df=pd.DataFrame()

df['Dataset Name']=Dataset_Name[1:589]
df['Dataset Type']=Data_Type[1:589]
df['Default Task']=Default_Task[1:589]
df['Attribute Type']=Attributes_Types[1:589]
df['No of instances']=Instances[1:]
df[' No of attribute']=Attributes[1:]
df['Year']=Year[1:]

df
```
