4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on
www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
# Activating the chrome browser
driver = webdriver.Chrome("chromedriver.exe")
```

```python
# getting the web page with the provided url
url = "https://www.flipkart.com/"
driver.get(url)
```

```python
search=driver.find_element_by_xpath("//input[@class='_3704LK']")
search.send_keys('reno 6 pro')
```

```python
btn=driver.find_element_by_xpath("//button[@class='L0Z3Pu']").click()
```

```python
SmartPhone_Name=[]
Brand_Name=[]
Url=[]

sp_name=driver.find_elements_by_xpath("//div[@class='_4rR01T']")
for i in sp_name:
    SmartPhone_Name.append(i.text)

bn=driver.find_elements_by_xpath("//div[@class='_4rR01T']")
for i in bn:
    Brand_Name.append(i.text[0:5])
    
url=driver.find_elements_by_xpath("//a[@class='_1fQZEK']")
for i in url:
    Url.append(i.get_attribute('href'))
```

```python
Model_No=[]
Color=[]
Price=[]

for i in Url:
    driver.get(i)
    m=driver.find_element_by_xpath("//div[@class='_1AtVbE col-12-12']//table[1]//tr[2]//td[2]//li[1]")
    Model_No.append(m.text)
    clr=driver.find_element_by_xpath("//div[@class='_1AtVbE col-12-12']//table[1]//tr[4]//td[2]//li[1]")
    Color.append(clr.text)
    prc=driver.find_element_by_xpath("//div[@class='_25b18c']//div[1]")
    Price.append(prc.text)
```

```python
dis=[]
res=[]
pros_type=[]
pros_core=[]
rom=[]
ram=[]
pr_cam=[]
sec_cam=[]
battery=[]

for i in Url:
    driver.get(i)
    driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']").click()
    
    model=driver.find_elements_by_xpath("//tr[@class='_1s_Smc row']")
    for n in model:
        text1=n.find_element_by_xpath("td[1]").text
        if text1=='Display Size':
            dis.append(n.find_element_by_xpath('td[2]').text)
        if text1=='Resolution':
            res.append(n.find_element_by_xpath('td[2]').text)
            
            
        try:
            if text1=='Processor Type':
                pros_type.append(n.find_element_by_xpath('td[2]').text)
        except:
            pros_type.append("-")
                                 
        if text1=='Processor Core':
            pros_core.append(n.find_element_by_xpath('td[2]').text)
        if text1=='Internal Storage':
            rom.append(n.find_element_by_xpath('td[2]').text)
        if text1=='RAM':
            ram.append(n.find_element_by_xpath('td[2]').text)
        if text1=='Battery Capacity':
            battery.append(n.find_element_by_xpath('td[2]').text)
        if text1=='Primary Camera':
            pr_cam.append(n.find_element_by_xpath('td[2]').text)
        if text1=='Secondary Camera':
            sec_cam.append(n.find_element_by_xpath('td[2]').text)
```

```python
df=pd.DataFrame()

df['Brand Name']=Brand_Name
df['Smartphone Name']=SmartPhone_Name
df['Model No.']=Model_No
df['Color']=Color
df['RAM']=ram
df['ROM']=rom
df['Primary Camara']=pr_cam
df['Secondary Camara']=sec_cam
df['Dispaly']=dis
df['Resoluation']=res
df['Processor']=pros_type
df['Processor Core']=pros_core
df['Battery Capacity']=battery
df['Price']=Price
df['Product URL']=Url

df
```
